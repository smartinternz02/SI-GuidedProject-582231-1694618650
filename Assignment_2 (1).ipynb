##Task 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

##Task 2

```python
data = pd.read_csv("/content/House Price India.csv")
data.head()
```

```python
data.columns
```

##Task 3  Univariate Analysis

```python
rcParams['figure.figsize'] = 12,8
plt.plot(data['living_area_renov'])
plt.title("Living Area Renovated(in m^2)")
plt.grid(True)
plt.show()
```

```python
rcParams['figure.figsize'] = 6,6
plt.boxplot(data.Price)
plt.title("Outliers in Prices")
plt.show()
```

##Task 2 Bi - Variate Analysis

```python
plt.scatter(data['condition of the house'],data['Built Year'])
```

```python
rcParams['figure.figsize'] = 10,8
plt.bar(data['Distance from the airport'],data['Price'])
plt.title("Effect of Airport on Price")
plt.xlabel('Distance from the airport')
plt.ylabel('Prices')
plt.grid(True)
plt.show()
```

```python
sns.jointplot(x=data['living area'],y=data['Price'],data=data)
```

##Task 3 Multivariate Analysis

```python
sns.pairplot(data)
```

```python
data.corr()
```

```python
rcParams['figure.figsize'] = 20,12
sns.heatmap(data.corr(),annot=True)
```
